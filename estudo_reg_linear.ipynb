## Regressão Linear

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = "svg"
```

#### Exemplificando com Numpy

* Instanciando os dados:

```python
X = np.array([0,1,2,3])
y = np.array([0,1,2,3])
```

* Definindo a função que calcula valores preditos, desde que possua alpha e beta:

```python
def predict(alpha, beta, X):
    return alpha*X + beta
```

* Definindo um alpha e um beta:

```python
alpha = 1
beta = 0

y = predict(alpha=alpha, beta=beta, X=X)
```

* Visualizando a reta criada pela predição:

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(X, y, c="b")
plt.plot(X, y, "r")
plt.text(0, 2.9, "Alpha={} | Beta={}".format(alpha,beta))
```

* Cálculo da melhor opção de alpha e beta:

```python
tamanho = len(X)
X_b= np.c_[np.ones((tamanho, 1)), X]
X_b
```

```python
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_best
```

```python
alpha = theta_best[1]
beta = theta_best[0]
```

```python
y = predict(alpha=alpha, beta=beta, X=X)

fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(X, y, c="b")
plt.plot(X, y, "r")
plt.text(0, 2.9, "Alpha={} | Beta={}".format(alpha,beta))
```

#### Utilizando dados aleatórios

```python
X = np.random.randint(4,20,(20,1))
y = (3 * X + np.random.randint(1,15,(20,1)))
```

```python
fig, ax = plt.subplots(figsize=(4,4))
plt.scatter(X,y,c="purple")
```

* Encontrando os melhores alpha e beta:

```python
tamanho = len(X)
X_b= np.c_[np.ones((tamanho, 1)), X]
X_b
```

```python
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_best
```

```python
alpha = float(theta_best[1])
beta = float(theta_best[0])
```

```python
y1 = predict(alpha=alpha, beta=beta, X=X)
```

* Verificando o gráfico com a reta que indica a função com melhor predição:

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(X, y, c="purple")
plt.plot(X, y1, "r")
```

#### Utilizando ScikitLearn para Regressão Linear

```python
from sklearn.linear_model import LinearRegression
LinearRegression = LinearRegression()
```

```python
X = np.random.randint(4,20,(20,1))
y = (3 * X + np.random.randint(1,15,(20,1)))

LinearRegression.fit(X, y)
```

```python
y1 = LinearRegression.predict(X)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(X, y, c="b")
plt.plot(X, y1, "r")
```

```python
print("Acurácia:",round(LinearRegression.score(X,y),3))
```
